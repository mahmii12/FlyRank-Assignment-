## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

**Question:** Which pages should be reviewed first for refresh, expansion, protection, pruning, or monitoring?

**Unit of analysis:** One content page (content_id), for a given client.

**Decision:** A content reviewer with limited time must choose which pages to spend their attention on first.

**Action:** Based on the recommendation, the reviewer refreshes, rewrites, expands, protects, or simply monitors a page.

**Cost of a wrong call:**
- False positive (flagged but page is actually fine): wastes reviewer time — a moderate cost, since review capacity is limited.
- False negative (a genuinely declining, high-demand page is missed): traffic loss goes unnoticed and unaddressed — potentially a larger cost.
- Because both errors matter, the ranking needs to be both precise at the top and not blind to real high-value opportunities.

**Why data/ML helps:** There are hundreds of thousands of content pages — far too many for manual review. A data-driven ranked score lets a human reviewer focus limited attention where it's statistically most likely to matter, instead of guessing or reviewing pages in arbitrary order.
